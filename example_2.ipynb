```python
big_2d_array_1 = big_array_1.reshape(4, -1, big_array_1.shape[-1])
```

```python
def gaussian_func(x, a, b, c):
    return a * np.exp(-(x - b)**2 / (2 * c**2))

def sinc_func(x, a, b, c):
    return a * np.sinc((x - b) / c)

def analyse(results_arr, f_linspace, custom_label: Optional[str] = 'Fine', chi_est: Optional[float] = 0.6 * MHz, fit_func: Optional[str] = 'gaussian'):
    supported_funcs = {'gaussian': gaussian_func, 'sinc': sinc_func}
    if fit_func not in supported_funcs.keys():
        raise ValueError(f'only these fit funcs are supported currently: {supported_funcs}')
    f_func = supported_funcs[fit_func]
    mean_res = np.mean(results_arr, axis=-1)
    abs_res = np.abs(mean_res)
    abs_g, abs_e = abs_res
    p0_g = [7.0, -0.5 * chi_est, 5.0e6]
    p0_e = [7.0, 0.5 * chi_est, 5.0e6]
    popt_g, pcov_g = curve_fit(f_func, f_linspace, abs_g, p0_g)
    popt_e, pcov_e = curve_fit(f_func, f_linspace, abs_e, p0_e)

    f_g = popt_g[1]
    f_e = popt_e[1]
    chi = np.abs(f_g - f_e)

    plt.plot(f_linspace/MHz, abs_res[0], label='g')
    plt.plot(f_linspace/MHz, gaussian_func(f_linspace, *popt_g), label=f'Fit g')
    plt.plot(f_linspace/MHz, abs_res[1], label=f'e')
    plt.plot(f_linspace/MHz, gaussian_func(f_linspace, *popt_e), label=f'Fit e')
    plt.axvline(x=f_g/MHz, color='green', label=f'freq g: {int(popt_g[1]/MHz*1e3)/1e3}MHz')
    plt.axvline(x=f_e/MHz, color='orange', label=f'freq e: {int(popt_e[1]/MHz*1e3)/1e3}MHz')
    plt.xlabel('Frequency (MHz)')
    plt.ylabel('Amplitude (A.U.)')
    plt.title(f'Frequency Spec {custom_label} Chi: {int(chi/MHz*1e3)/1e3}MHz')
    plt.legend(bbox_to_anchor=(1.0, 1.0))
    plt.show()

chi_est = 0.6 * MHz
fit_func = 'gaussian'

analyse(big_2d_array_1[:2], rr_lins_fine, custom_label='Fine', chi_est=chi_est, fit_func=fit_func)
analyse(big_2d_array_1[2:], rr_lins_large, custom_label='Rough', chi_est=chi_est, fit_func=fit_func)
```
